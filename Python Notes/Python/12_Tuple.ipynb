# **Tuple** :

### `defination :` Tuple are used to store multiple items in a single varible.

> ### Tuple is one of 4 built-in data types in Python used to store collection of data, the other 3 are "list", "set" and "dictionary", all with different qualities and usages.

> ### A tuple is a collection which is `ordered` and `unchangeable`.

> ### Tuples are written with round brackets ().

```python
tupleData = ("mohan", "sohan", "anuj", "gopal")
print(tupleData)
print(type(tupleData))
```

## **Tuple Items** : 

> ### Tuple itmes are `ordered`, `unchangeable`, and allow `duplicate` values.

> ### Tuple items are indexed, first item has index [0], the second item has index [1] etc.

### `Ordered`

> #### When we say that tuples are ordered, it means that the items have a defined order, and that order will not change.

### `Unchangeable`

> #### Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created.

### `Allow Duplicates`

> #### Since tuple are indexed, they can have items with the same value.

```python
tupleItems = ('mango', 'grapes', 'papaya', 'mango', 'pineapple')
print(tupleItems)
```

## Tuple Length : 

> #### To determine how many items a tuple has, use the `len()` function.

```python
tupleItems = ('mango', 'grapes', 'papaya', 'mango', 'pineapple')
print(len(tupleItems))
```

## **Access Tuple Items** : 

```python
# we can access tuple items by referring to the index number, inside square brackets.

tupleData = ("mohan", "sohan", "anuj", "gopal")
print(tupleData[2])
```

```python
# using negative indexing means start from the end.
# -1 means the last element, -2 means the second last element etc.

tupleData = ("mohan", "sohan", "anuj", "gopal")
print(tupleData[-2])
```

## `Range` of Indexes : 

> ### We can specify a range of indexes by specifying where to start and where to end the range. When specifying a range, the return value will be a new tuple with the specified items.

```python
rgTuple = ("anupam", "ayushi", "aaravh", "mohan", "madhav", "mahesh", "madhu")

print(rgTuple[2:5])

print(rgTuple[:4])

print(rgTuple[3:])

print(rgTuple[:])
```

```python
print(rgTuple[-4:-1])
```

## **Check** if item exists :

> ### To determine if a specified item is present in a tuple use the `in` keyword.

```python
anyTuple = ("apple", "papaya", "ayushi", "mohan", "anupam")

if "apple" in  anyTuple:
    print("apple is present in the tuple")
else:
    print("apple is not present in the tuple")
```

## **Update Tuple** : 

## Change Tuple Values : 

> ### Once a tuple is created, you can not change its values. Tuples are `unchangeable`, or `immutable` as it also is called.

> ### But there is a workaround. You can convert the tuple into a list, change the list, and convert the list back into a tuple.

```python
# create a tuple:
studentNames = ("Vanshika", "Ankit", "Deepak", "Vineet")

# Step 1 -> convert tuple into list:
listStdNames = list(studentNames)

# Step 2 -> update list items:
listStdNames[0] = "Vansh"

# Step 3 -> convert back list into tuple:
tupleStdNames = tuple(listStdNames)

# print updated tuple items:
print(tupleStdNames)
print(type(tupleStdNames))
```

## Add Items : 

> ### Since tuples are immutable, they do not have a built-in `append()` method, but there are other ways to add items to a tuple.

```python
# 1. Convert into a list ->

studentNames = ("Vanshika", "Ankit", "Deepak", "Vineet")
stdList = list(studentNames) # convert tuple into list
stdList.append("Ayushi") # append new data
stdList.append("Girdhar") # append new data
stdTuple = tuple(stdList) # convert back list into tuple
print(stdTuple)
```

```python
# 2. Add tuple to a tuple ->

studentNames = ("Vanshika", "Ankit", "Deepak", "Vineet")
newName = ("Anuradha",) # create a single item tuple
studentNames += newName # add tuple to an another tuple
print(studentNames)
```

## Remove Items :

### `You can not remove items in a tuple.`

```python
studentNames = ("Vanshika", "Ankit", "Deepak", "Vineet")
stdList = list(studentNames) # convert tuple into list
stdList.remove("Vineet") # remove an item
studentNames = tuple(stdList) # convert back list into tuple
print(studentNames)
```

```python
# here, i am using "del" keyword:
studentNames = ("Vanshika", "Ankit", "Deepak", "Vineet")
del studentNames
print(studentNames)
```

## **Unpack Tuples** : 

> ### ***When we create a tuple, we normally assign values to it. This is called `packing` a tuple.***

> ### ***But, in Python, we are also allowed to extract the values back into variables. This is called `unpacking`.***

```python
# create a tuple:
v_names = ("Mohan", "Girdhar", "Govind", "Damodar")

# unpacking the tuple items:
(k1, k2, k3, k4) = v_names

# print unpacked items:
print(f"My first fav. krs. name is {k1}")
print(f"My second fav. krs. name is {k2}")
print(f"My third fav. krs. name is {k3}")
print(f"My fourth fav. krs. name is {k4}")

# note: the number of variables must match the number of values in the tuple, if not you match use an asterisk to collect thre remaining values as a list.
```

## Using Asterisk* :

```python
v_names = ("Mohan", "Girdhar", "Govind", "Damodar")

# unpacking the tuple items:
(k1, *k2) = v_names

# print unpacked items:
print(k1)
print(k2)

# here, iterate the "k2" list data:
for i in k2:
    print(i)

# note: here, k1 as a variable kaam kar raha hai aur k2 as a list kaam kar raha hai, k2 me baaki ke bache hue items list ki format me store ho jayenge.
```

```python
v_names = ("Mohan", "Girdhar", "Govind", "Damodar", "Madhuvan", "Bhaskar")

# unpacking the tuple items:
(k1, *k2, k3, k4) = v_names

# print unpacked items:
print(k1)
print(k2)
print(k3)
print(k4)
```
