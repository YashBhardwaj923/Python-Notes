# **SET**

> ### Set are used to store multiple items in a single variable.
> ### A Set are collection which is `unoredered`, `unchangeble`, and `unindexed`.
> ### Set items are `unchangeble`, but you can remove items and add new items.
> ### Sets are written with curly brackets {} .

```python
# example :
sets = {"mohan", "sohan", "anuj", "gopal"}
print(sets)
```

> ### `note:` Sets are unordered, so you cannot be sure in which order the items will appear.

## **Set items are `unordered`, `unchangeable` and `do not allow duplicate` values**.

## ***`Unordered` :***

* ### Unordered means that the items in a set do not have a defined order.
* ### Set items can appear in a different order every time you use them, and cannot be referred to by index or key.

## ***`Unchangeable` :***

* ### Set items are unchangeable, meaning that we cannot change the items after the set has been created.

## ***`Duplicates Not Allowed` :***

* ### Sets cannot have two items with same value.

```python
d_items = {"apple", "mango", "grapes", "apple"}
print(d_items)
```

```python
d_items = {True, False, 1, 2}
print(d_items)
```

## **Get the Length of a Set :**

```python
names = {"mohan", "sohna", "anuj", "gopal"}
print(len(names))

print(type(names))
# here, type() function, for get data type 
```

## **`set()` Constructor :**

```python
any_set = set(("apple", "mango", "grapes", "appple"))
print(any_set)
```

# **Access Set Items**

> #### We cannot access items in a set by referring to an index or a key.
> #### BUT we can loop through set items using a '`for`' loop or ask if a specified value is present in a set, by using the '`in`' keyword.

```python
names = {"mohan", "sohna", "anuj", "gopal"}

# using `for` loop :
for name in names:
    print(name)
```

```python
print("sohna" in names)
```

#### `note`: Once a set is created, we cannot change its items, but we can add new items.

# **Add Set Items**

### `add()` method:

```python
fnds = {"ayushi", "mohan", "sohan"}
fnds.add("gopal")
print(fnds)
```

### `update()` method: **for add new set.**

```python
fnds1 = {"ayushi", "mohan", "sohan"}
fnds2 = {"anupam", "anurag"}
fnds1.update(fnds2)
print(fnds1)
```

# **Remove Items**

### `remove()` or `discard()` method:

```python
fnds = {"ayushi", "mohan", "sohan"}
fnds.remove("ayushi")
print(fnds)

# fnds.remove("gopal")
# print(fnds)
```

```python
fnds = {"ayushi", "mohan", "sohan"}
fnds.discard("mohan")
print(fnds)

# fnds.discard("gopal")
# print(fnds)
# note: if the item to remove does not exist, `discard()` method will not raise an error.
```

### `pop()` method: **remove random items**

```python
fnds = {"ayushi", "mohan", "sohan"}
fnds.pop()
print(fnds)

removefnd = fnds.pop()
print(fnds)
print(removefnd)
```

### `clear()` method:

```python
fnds = {"ayushi", "mohan", "sohan"}
fnds.clear()
print(fnds)
```

### `del` keyword:

```python
fnds = {"ayushi", "mohan", "sohan"}
del fnds
print(fnds)
```
