# **DICTIONARY**

##### `defination` : Dictionary are used to store data values in `key:value` pairs.

* A dictionary is a collection which is `ordered`, `changeable` and `do not allow duplicates`.

* Dictionaries are written with curly brackets `{}`.

```python
disc = {
    "id": 1,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}
print(disc)
```

## ***Dictionary Items :***

* Dictionary items are `ordered`, `changeable`, and `do not allow duplicates`.

```python
disc = {
    "id": 2,
    "name": "Mohan Kumar",
    "age": 31,
    "city": "Mohan Nagar",
    "city": "Hathras" # duplicate value (re-assign same name key)
}
print(disc)
```

## ***Access Dictionary Items :***

```python
# 1. we can access the items of a dictionary by referring to its key name, inside square brackets [].

disc = {
    "id": 2,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

print(disc["name"])
print(disc["age"])
```

```python
# 2. There is also a method called `get()` that will give you the same result.

disc = {
    "id": 3,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

print(disc.get("name"))
```

### `.keys()` method will return a list of all the keys in the dict.

```python
disc = {
    "id": 4,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

keyss = disc.keys()
print(keyss)
```

### `.values()` method will return a list of all the values in the dict.

```python
disc = {
    "id": 4,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

valuess = disc.values()
print(valuess)
```

### `.items()` method will return each item in a dict., as tuple in a list.

```python
disc = {
    "id": 4,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

itemss = disc.items()
print(itemss)
```

### `in` keyword determine if a specified key is present in a dict use the `in` keyword.

```python
disc = {
    "id": 5,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

# i.
if "id" in disc:
    print("id is present")
else:
    print("id is not present")

# ii.
for id in disc:
    print(id)
    if id=="id":
        break
```

## ***Change Dictionary Items :***

```python
# 1. we can change the value of a specific item by referring to its key name.

disc = {
    "id": 5,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

disc["city"] = "Hathras"
print(disc)
```

#### `update()` method will update the dict with the items from the given argument.

```python
disc = {
    "id": 5,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

disc.update({"name": "T.Sharma"})
print(disc)
```

## ***Add Dict New Items :***

```python
disc = {
    "id": 6,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

disc["gender"] = "Male"
print(disc)
```

```python
disc = {
    "id": 7,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

disc.update({"country": "India"})
print(disc)
```

## ***Remove Dict. Items :***

#### `.pop()` method removes the item with the specified key name.

```python
disc = {
    "id": 8,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

disc.pop("age")
print(disc)
```

#### `.popitem()` method removes the last item.

```python
disc = {
    "id": 9,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

disc.popitem()
print(disc)
```

#### `del` keyword.

```python
disc = {
    "id": 10,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

del disc["name"]
print(disc)

del disc
print(disc)
```

#### `.clear()` method empties the dict.

```python
disc = {
    "id": 10,
    "name": "Tushar Sharma",
    "age": 28,
    "city": "New Delhi"
}

disc.clear()
print(disc)
```

## ***Loop Dictionaries :***

```python
l_dict = {
    "name": "Mohan Kumar",
    "city": "Noida",
    "skills": ["Python", "Django", "Front-End"]
}

for x in l_dict:
    print(x) # keys
    print(l_dict[x]) # values

# here, x - stores dict keys.
```

#### `.keys()` and `.values()`

```python
for i in l_dict.keys():
    print(i)
# here, "i" variable me dict ke keys store ho rahi hain.

for i in l_dict.values():
    print(i)
# here, "i" variable me dict ke values store ho rahi hain.
```

#### `.items()`

```python
for x, y in l_dict.items():
    print(f"{x} : {y}")

# using break statement:
for x, y in l_dict.items():
    print(f"{x} : {y}")
    if x == "city":
        break
```

## ***Copy Dictionaries :***

```python
main = {
    "id": 1,
    "name": "Mohan Kumar",
    "age": 31,
    "city": "Mohan Nagar",
    "skills": ["Python", "Django", "Front-End"]
} # 100011010
```

```python
user = main # 100011010
# here, maine "user" naam ke variable me "main" dict ke data ko copy kar diya.

# after copy dicts, change in user's data:
user["name"] = "Tushar Sharma"
print(f"user dict : {user}")
print(f"main dict : {main}")
```

#### `.copy()`

```python
main = {
    "id": 1,
    "name": "Mohan Kumar",
    "age": 31,
    "city": "Mohan Nagar",
    "skills": ["Python", "Django", "Front-End"]
} # 1000110101

user = main.copy() # 1001010101

# after copy dicts, change in user's data:
user["name"] = "Tushar Sharma"
user["city"] = "San Francisco"
print(f"user dict : {user}")
print(f"main dict : {main}")
```

#### `dict()`

```python
main = {
    "id": 1,
    "name": "Mohan Kumar",
    "age": 31,
    "city": "Mohan Nagar",
    "skills": ["Python", "Django", "Front-End"]
}

user = dict(main)

# after copy dicts, change in user's data:
user["name"] = "Tushar Sharma"
user["city"] = "San Francisco"
print(f"user dict : {user}")
print(f"main dict : {main}")
```

#### ***How to Iterate List in Dicts :***

```python
main = {
    "id": 1,
    "name": "Mohan Kumar",
    "age": 31,
    "city": "Mohan Nagar",
    "skills": ["Python", "Django", "Front-End"]
}

skills = main["skills"]

for skill in skills:
    print(skill)
```

## ***Nested Dictionaries :***

```python
nested_dict = {
    "std1": {
        "id": 1,
        "name": "Mohan",
        "city": "Delhi"
    },
    "std2": {
        "id": 2,
        "name": "Rahul",
        "city": "Mumbai"
    },
    "std3": {
        "id": 3,
        "name": "Tushar",
        "city": "New Delhi"
    }
}

print(nested_dict)

print(nested_dict["std2"]["city"])
print(nested_dict["std3"]["name"])
```
