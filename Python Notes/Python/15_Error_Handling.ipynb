# **Different Types of Errors**

> Errors are the problems in a program due to which the program will stop the execution. On the other hand,
 exceptions are raised when some internal events occur which changes the normal flow of the program.

> Two types of Error occurs in python.
*    1. Syntax Errors
*    2. Logical Errors (Exceptions)

### ***Syntax Errors*** : 

```python
age = 18
if age < 18;
    print("Yes")
else:
    print("No")

# note: syntax errors resolve nahi hoti hain.
```

### ***Logical Errors (Exceptions)*** : 

> There are different types of logical errors:
*    i.    ZeroDivisionError:
*    ii.   NameError:
*    iii.  TypeError:
*    iv.   ValueError:
*    v.    IndexError:
*    vi.   KeyError:
*    vii.  ModuleNotFoundError: 
*    viii. ImportError:

#### `ZeroDivisionError`

```python
a = 0
divide = 1/a
print(divide)
# output - ZeroDivisionError: division by zero
```

#### `NameError`

```python
print(x) # here, x is not defined in this scope
# output - NameError: name 'x' is not defined

# note: yadi kisi bhi variable ko define karne se pahle use kar liya jaye to... output "variable" is not defined aata hai.
```

#### `TypeError`

```python
l = 10
m = "20"
sum = l + m
print(sum)
# output - TypeError: unsupported operand type(s) for +: 'int' and 'str'
```

#### `ValueError`

```python
hi = int("eleven")
print(hi)
# output - ValueError: invalid literal for int() with base 10: 'eleven'
```

#### `IndexError`

```python
names = ["mohan", "sohan", "anuj", "gopal"]
print(names[4])
# output - IndexError: list index out of range
```

#### `KeyError`

```python
user = {
    "id": 1,
    "name": "Tushar Sharma",
    "age": 28
}
print(user["address"])
# output - KeyError: 'address'
```

#### `ModuleNotFoundError`

```python
import mathhs

#  ModuleNotFoundError: No module named 'mathhs'
```

# **Exceptional Handling**

> The '`try`' block lets you test a block of code for errors.

> The '`except`' block lets you handle the error.

> The '`else`' block lets you execute code when there is no error.

> The '`finally`' block lets you execute code, regardless of the result of the try- and except blocks.

```python
# 1.
print("Start Code")

try:
    print(data)
except Exception as e:
    print(e)

print("Important Code")
print("End Code")
```
