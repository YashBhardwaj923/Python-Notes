# Variables :

> Variables is a **container** for storing data/values.

```python
# initialization a variable :
x = "Hello"

"""
explaination :-

x  --> variable name
=  --> assignment operatot
"Hello"  --> value assigned to the variable
"""

# print a variable value using print() method :
print(x)
```

### Variable **re-assign**, **re-declare**, **re-initialize** :

```python
name = "Gopal Verma" # initialization variable
name = "Anshuman Joshi" # re-assign/re-declare/re-initialize same name variable

print(name) # print name variable
```

### How to define a variable name ?

```python
first_name = "Tushar"
last_name = "Sharma"

num = 67
strr = "Hello Shree!"
arr = ["Hello", "Tushar", 56, 23]

print(first_name)
print(last_name)
print(num)
print(strr)
print(arr)
```

### How to declare multi word variable name ?

* camel case
* pascal case
* snake case

```python
# my friend name :

myFriendName = "Tushar Sharma" # camel case
print(myFriendName)

MyFriendName = "Aushman Joshi" # pascal case
print(MyFriendName)

my_friend_name = "Anurag Bhardwaj" # snake case
print(my_friend_name)

# my-friend-name = "Mohan Kumar" ❌
# my frined name = "Madhaw Joshi" ❌
# 1name = "Goapl" ❌
```

### Get the Data Type :

> If we are geting type of data, using **type()** function in Python.

```python
my_name = "Yash Bhardwaj"
print(my_name)
print(type(my_name))
```

### String Qutation :

> Single ( '' ) or Double ( "" ) qutations are same.

```python
str = "Mohan's ball"
sen = 'My name is Tushar Sharma'

print(str)
print(sen)
```

### Case-Sensitive Variables :

```python
age = 14
Age = 28
AGE = 54

print("age : ", age)
print("Age : ", Age)
print("AGE : ", AGE)
```

### Many Values to Multiple Values :

```python
a, b, c = 10, 20, 30
print(a)
print(b)
print(c)

name, city, email = "Tushar Sharma", "Gurugram", "sharma321@gmail.com"
print(name)
print(city)
print(email)

print(name, city, email)
```

### One Value to Multiple Variables :

```python
a = b = c = 34
print(a)
print(b)
print(c)

"""
a = 34
b = a
c = b 
"""
```
