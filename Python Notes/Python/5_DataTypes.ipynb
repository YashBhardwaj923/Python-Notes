# Data Types :

> Different types of data type in Python :

1. **Numeric Data Type** ---> (int, float, complex)
2. **Text Data Type** ---> (str)
3. **Boolean Data Type** ---> (True, False)
4. **Sequence Data Type** ---> (list(mutable), tuple(immutable))
5. **Set Data Type** ---> (set) [Set is an unordered collection of data types that is iterable, mutable, and has no duplicate elements.]
6. **Mapped Data Type** ---> (dict)

### 1. Numeric Data Type :-

#### ***int*** -

```python
num = 23
print(num)
print(type(num))

# int() function :-
inum = int(43) # here, i am using "int()" function for declare a intiger value
print(inum)
print(type(inum))
```

##### ***float*** -

```python
f_num = 45.5
print(f_num)
print(type(f_num))

# float() function :-
fnum = float(32.6) # here, i am using "float()" function for declaring the float value
print(fnum)
print(type(fnum))
```

##### ***complex*** -

```python
c_data = 1j + 2
print(c_data)
print(type(c_data))

# complex() function :-
cdata = complex(3, 4) # here, i am using "complex()" function for declaring the complex number
print(cdata)
print(type(cdata))
```

### 2. Text Data Type :

##### ***str*** -

```python
my_name = "Tushar Sharma"
print(my_name)
print(type(my_name))


# str() function :-
sname = str("Shree") # here, i am using "str()" function for declaring the string value
print(sname)
print(type(sname))
```

### 3. Boolean Data Type :

##### ***True*** & ***False*** -

```python
tr = True
fl = False
print(tr)
print(type(tr))
print(fl)
print(type(fl))

# bool() function :-
btr = bool(10) # here, i am using "bool()" function for declaring the boolean value
bfl = bool(0) # here, i am using "bool()" function for declaring the boolean value
print(btr)
print(type(btr))
print(bfl)
print(type(bfl))
```

### 4. Sequence Data Type :

##### ***list*** -

```python
# declaring a list :
friends = ['tushar', 'rahul', 'anuj', 'mohan', 'gopal']
print(friends)  # printing the list
print(type(friends))

# access the list elements :
print(friends[0])
print(friends[1])

# update list elements :
friends[3] = "anurag"
print(friends)

# list  --->>  ['tushar', 'rahul', 'anuj', 'mohan', 'gopal']
# index --->>      0         1        2       3        4
# length -->>      1         2        3       4        5
```

```python
# store multiple data type in a single variable using list :

data = ["mohan", 'shree', True, [1, 2, 3]]
print(data)
print(type(data))

data[3][0] = 12
print(data)
```

```python
# declare a list using "list()" function :
city = list(["delhi", "agra", "noida", "mathura"])
print(city)
print(type(city))
```

*index formula* :- **lenght - 1 = index**

##### ***tuple*** -

```python
# declare a tuple data :
data = ("mohan", "sohan", "anurag")
print(data)
print(type(data))

# access tuple data :
print(data[0])
print(data[1])

# update tuple data :
# tuples are immutable, so we can't update the tuple data.
# data[0] = "aushman"
# print(data)
```

```python
# using "tuple()" function :

data = tuple(["tushar", "rahul", "anuj", "mohan", "gopal"])
print(data)
print(type(data))
```

### 5. Set Data Type :

```python
# declare a set data :
data = {"tushar", "mohan", "sohan", "anurag"}
print(data)
print(type(data))

# access set data :
# print(data[0]) # it will throw an error because set is unordered collection of data types and indexing is not possible.

# update set data :
data.add("GOPAL")
print(data)

# no duplicate elements support :
data = {"tushar", "sohan", "rohan", "tushar"}
print(data)
```

```python
m_data = {"sohan", 29, "rohan", 86}
print(m_data)
```

### 6. Mapped Data Type :

##### ***dict*** -

```python
# declare a dict :
user_1 = {
    "name": "Tushar",
    "age": 27,
    "address" : "Hathras"
}
print(user_1)
print(type(user_1))

# access dict data :
print(user_1["name"])
print(user_1["age"])

# update dict data :
user_1["age"] = 28
print(user_1)
```
