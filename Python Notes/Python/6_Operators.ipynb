# Operators (OOO)

> ***Operator Operates the Operands***.

* **Operator** = +, -, *, /, %...etc.
* **Operates** = operation with two or more numbers or data.
* **Operand**  = the numbers or data that are operated on.

#### There are different types of operators in **Python** :

* 1. **Arithmetic** Operator
* 2. **Assignment** Operator
* 3. **Comparison** Operator
* 4. **Logical** Operator
* 5. **Indentity** Operator
* 6. **Membership** Operator
* 7. **Bitwise** Operator

#### **4. Logical Operators :**

```python
"""Operator	Description                             	                    Example	
     and 	    Returns True if both statements are true	                     x < 5 and  x < 10	
     or	    Returns True if one of the statements is true	                x < 5 or x < 4	
     not	    Reverse the result, returns False if the result is true       not(x < 5 and x < 10)"""
```

```python
# "and" operator :
count = 25
print(count > 18 and count < 30)
print(count > 26 and count < 30)
```

```python
# "or" operator :
count = 25
print(count > 18 or count < 30)
print(count > 26 or count < 30)
print(count > 26 or count < 18)
```

```python
# "not" operator :
count = 20
print(not count >= 18)
```

#### **5. Identity Operators :**

```python
"""Operator	    Description	                                                Example
    is 	        Returns True if both variables are the same object	        x is y	
    is not	    Returns True if both variables are not the same object	    x is not y"""
```

```python
a = 10
b = 10
print(a is b)
```

```python
print(a is not b)
```

#### **6. Membership Operator :**

```python
"""Operator	    Description	                                                                        Example
    in 	        Returns True if a sequence with the specified value is present in the object	    x in y	
    not in	    Returns True if a sequence with the specified value is not present in the object	x not in y"""
```

```python
# (in)
names = ["Mohan", "Sohan", "Rahul", "Gopal", "Anuj"]

print("mohan" in names)
print("Mohan" in names)
```

```python
print("mohan" not in names)
print("Gopal" not in names)
```

#### **7. Bitwise Operator :**

```python
'''Operator	   Example   Name  
    &	        x & 3	 AND    	
    |	        x | 3	 OR	
    ^	        x ^ 3    XOR'''	

""" FORMULAS :->     ...16 + 8 4 2 1 """
```

```python
# AND (&) :->
a = 7
b = 9

# here, i am using "bin()" function for converting decimal to binary value.
print(bin(a))
print(bin(b))

print(a & b)
```

```python
# OR (|) :-->
x = 34
y = 23

print(bin(x))
print(bin(y))

print(x | y)
```

```python
# XOR (^) :-->
l = 43
m = 32

print(bin(l))
print(bin(m))

print(l ^ m)
```
