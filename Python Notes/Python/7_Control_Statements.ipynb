# Control Statements :

> There are different types in Python : 

* 1. **Conditional statements** :
    *   if
    *   if...else
    *   if...elif...else
* 2. **Looping statements** :
    *   while
    *   for
    *   break
    *   continue

### 1. Conditional Statements :

##### **i. ***"if"*** statement** : _1 condition and 1 statement_

```python
# age = int(input("enter your age : "))
age = 18

if age > 18:
    print("You are eligible for voting.")
```

##### **ii. ***"if...else"*** statement** : _1 condition and 2 statement_

```python
age = 12

if age >= 18:
    print("You are eligible for voting.")
else:
    print("You are not eligible for voting.")
```

##### **iii. ***"elif"*** statement** : 1+ conditions and 1+ statements

```python
age = int(input("enter your age please... : "))

if age >= 18 and age <= 39:
    print("han ji sir, aap vote de sakte ho.")
elif age >= 40 and age <= 80:
    print("aap to jarur jarur vot de sakte ho... voting vale din, chutti mat manana...")
else:
    print("bhai aisa hai, yahan bacchon ko vote dena alow nahi hai, pahle jao papa se puch ke aao.")
```

##### **iv. ***"nested if else"*** statement** : 

```python
time = int(input("enter the time (0-12) : "))

if time >= 0 and time < 12:
    zone = input("enter zone (am & pm)")

    if zone == "am":
        if time >= 0 and time < 12:
            print("Good Morning@...🌄")
    elif zone == "pm":
        if (time == 12) or (time >= 1 and time <= 4):
            print("Good Afternoon@...🌞")
        elif time > 4 and time <= 9:
            print("Good Evening@...🌇")
        elif time > 9 and time < 12:
            disc = input("enter your type (meet or sleep) : ")

            if disc == "meet":
                print("Good Evening----🤝")
            elif disc == "sleep":
                print("Good Night----😴")
            else:
                print("please enter your correct type 'meet' or 'sleep'.")
    else:
        print("please enter your correct zone 'am' or 'pm'.")
else:
    print("time is incorrect!")
```

### 2. Looping Statements :

##### **i. ***"while"*** statement** :

> entry control loop.

```python
count = 0

while count <= 10:
    print(count)
    count += 1
```

```python
twoTable = 2

while twoTable <= 20:
    print(twoTable)
    twoTable += 2
```

```python
fiveTable = 5

while fiveTable <= 50:
    print(fiveTable)
    fiveTable += 5
```

```python
table = int(input('enter table number : ')) 
count = int(input("enter count : ")) 

while count <= table*10: 
    print(count)
    count += table
```

```python
ipt = int(input('enter num : '))

while ipt <= ipt*10:
    print(ipt)
    ipt += ipt
```

```python
mala = 1

while mala <= 108:
    print("Radhey Radhey ❣️")
    mala += 1
```

```python
naam = "Shree Radha ❣️"
mala = 1

while mala <= 108:
    print(naam)
    mala += 1
```

##### **ii. ***"for"*** loop** :

```python
sent = "Hello Shree!@"

for x in sent:
    print(x)
```

```python
arr = ["anupam", "ayushi", "anurag", "mohan"]

for i in arr:
    print(i)
```

##### **iii. ***"break"*** statement** :


```python
arr = [1, 2, 3, 4, 5, 6, 7, 8]

for i in arr:
    print(i)
    if i == 5:
        break
```

```python
arr = [1, 2, 3, 4, 5, 6, 7, 8]

for i in arr:
    if i == 5:
        break
    print(i)
```

```python
count = 1

while count <= 5:
    print(count)
    count += 1
    if count == 3:
        break
```

```python
count = 1

while count <= 5:
    if count == 3:
        break
    print(count)
    count += 1
```

##### **ii. ***"continue"*** statement** :


```python
num = 1

while num <= 10:
    if not num % 2 == 0:
        num += 1
        continue
    print(num)
    num += 1
```

```python
nums = 0

while nums <= 10:
    if (nums % 2 == 0):
        print(nums)
    nums += 1
```
