# Functions

> Python function is a block of code, that runs only when it is called. It is programmed to return the specific task. You can pass values in functions called parameters. It helps in performing repetitive tasks.

i.  increase code `readability` 
ii. increase code `reusability`

#### Types of Functions :-
* `Built-in` function
* `user-defined` function
* `lambda` functions


#### 1. `Built-in` Functions :

Python provides various built-in functions that can be used to perform various tasks. Some of them are:

* `print()` : This function is used to display the output on the console.
* `type()` : This function is used to get the type of a variable.
* `len()` : This function is used to get the length of a string or list.
* `range()` : This function is used to generate a sequence of numbers.
* other functions `int`, `str`, `complex`, `bool`, `float`, `list`, `tuple`, `set` and `disc` etc...

#### 2. `User-define` Functions :

```python
# HOW TO CREATE A FUNCTION :->

def funs():
    print("Hello Shree!")

# CALLING/ INVOKING A FUNCTION :-
funs()
```

##### Explain the function :

here,...
* `def` is a function keyword. `funs` is a function name.
* The parentheses `()` are used to define the parameters of the function.
* The colon `:` is used to separate the function definition from the code inside the function.
* `def funs():` function declaration.
* `after ':' with indantation` is a function body/ excutable part.
* `print("Hello Shree!")` is the code inside the function.
* `funs()` is a function call/ invocation. It calls the `funs()` function.

```python
# HOW TO USE RETURN STATEMENT AND WHY ?

def name():
    return "Ayushi"

result = name()
print(result)
```

#### **Parameters** & **Arguments**

```python
# PARAMETERS AND ARGUMENTS :->

def sum(a, b):
    return a + b

result = sum(11, 33)
print(result)
```

here,...

* `def sum(a, b):` function definition with two parameters `a` and `b`.
* `a` and `b` is two parameters.
* `return a + b` is the return statement which returns the sum of `a` and `b`.
* `result = sum(11, 33)` is a function call/ invocation with arguments `11` and `33`.
* `11` and `33` is two arguments.
* `print(result)` prints the result of the function call.

##### **Que**. Write a Function for check if the given number is `even` or `odd`.

```python
def findEvenOdd(x):
    if x % 2 == 0:
        return "Even"
    else:
        return "Odd"
    
num = int(input("enter any decimal number : "))

result = findEvenOdd(num)
print(f"{num} is {result}.")
```

#### **Que** Write a calculator function.

```python
num1 = int(input("enter 1st number : "))
num2 = int(input("enter 2nd number : "))
oper = input("enter operator symbol +, -, *, / : ")

def calculator(num1, num2, operation):
    if operation == "+":
        return num1 + num2
    elif operation == "-":
        return num1 - num2
    elif operation == "*":
        return num1 * num2
    elif operation == "/":
        if num2 == 0:
            return "Error! Division by zero"
        else:
            return num1 / num2
    else:
        return "Error! Invalid operation"

result = calculator(num1, num2, oper)
print(f"{num1} {oper} {num2} = result is {result}.")
```

#### **Default Parameter**

```python
def sum(a, b=10):
    return a + b

result = sum(11, 56)
print(result)

# note: here, b=10 is a default parameter.
```

```python
def demo(x, y, z=30):
    return x + y + z

result = demo(10, 20)
print(result)

# note: here, z=30 is a default parameter.
```

#### ***arg**

```python
def some(l, m, n, *arg):
    print(l)
    print(m)
    print(n)
    for i in arg:
        print(i)

some(10, 20, 30, 40, 50, 60, 70 ,80)
```

#### **Named arguments**

```python
def info(name, age, city):
    return f"Name is : {name}.\nAge is {age}.\nCity is {city}."

result = info(age=34, city="Delhi", name="Rahul Sharma")
print(result)
```

#### **Function within Functions**

```python
def outer():
    x = "outer func var"

    def inner():
        y = "inner func var"
        print(y)

    inner()

    return x
   
result = outer()
print(result)
```

#### 3. `lambda` Functions :

```python
# define 'lambda' function :
hello = lambda : print("Hello Shree")

# call the 'lambda' function
hello()
```

##### Explaination:

* `hello` is a name of the function/ variable.
* `lambda` is a anonymous function keyword, it's define the anonymous function.
* `:` a colon for starts a function block.
* `print("Hello, World!")` is a function body, it's print the string.
* `hello()` call the function.

#### `lambda` with **parameters** statement:

```python
add = lambda a,b : print(a +b)

a = int(input("enter f_num : "))
b = int(input("enter s_num : "))
add(a, b)
```

```python
# default parameter with lambda function:

sum = lambda x,y=30 : print(x + y)
sum(10)
```

##### `lambda` with **return** statatement:

```python
sent = lambda pr : pr
output = sent("Hello Shree")
print(output)
```
