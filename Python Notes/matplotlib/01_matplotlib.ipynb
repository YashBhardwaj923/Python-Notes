## **`Create Labels for a Plot`**

> #### With Pyplot, we can use the `xlabel()` and `ylabel()` functions to set a label for the x and y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

plt.plot(x, y)

plt.xlabel("company rate")
plt.ylabel("company sales")

plt.show()
```

### **Add Title**:

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

plt.plot(x, y)

plt.title("Comany Sales vs Rate")
plt.xlabel("company rate")
plt.ylabel("company sales")

plt.show()
```

### **Set font properties for title and labels:**

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

font1 = {'family': 'serif', 'color': 'green', 'size': 15}
font2 = {'family': 'Cursive', 'color': 'red', 'size': 10}

plt.title("Comany Sales vs Rate", fontdict=font1)
plt.xlabel("company rate", fontdict=font2)
plt.ylabel("company sales", fontdict=font2)

plt.plot(x, y)

plt.show()
```

`font types:-`
* serif
* sans-serif
* monospace
* cursive
* fantasy

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

font1 = {'family': 'serif', 'color': 'green', 'size': 15}
font2 = {'family': 'serif', 'color': 'red', 'size': 10}

plt.title("Comany Sales vs Rate", fontdict=font1, loc='left') # here, i am using "loc=" function to arrange the title (left, right or center)
plt.xlabel("company rate", fontdict=font2)
plt.ylabel("company sales", fontdict=font2)

plt.plot(x, y)

plt.show()
```

## `Add Grid Lines to a Plot`

> Here, we can use the `grid()` function to add grid lines to the plot.

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

font1 = {'family': 'serif', 'color': 'green', 'size': 15}
font2 = {'family': 'serif', 'color': 'red', 'size': 10}

plt.title("Comany Sales vs Rate", fontdict=font1)
plt.xlabel("company rate", fontdict=font2)
plt.ylabel("company sales", fontdict=font2)

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

font1 = {'family': 'serif', 'color': 'green', 'size': 15}
font2 = {'family': 'serif', 'color': 'red', 'size': 10}

plt.title("Comany Sales vs Rate", fontdict=font1)
plt.xlabel("company rate", fontdict=font2)
plt.ylabel("company sales", fontdict=font2)

plt.plot(x, y)

plt.grid(axis='x')
# show grid only 'x' axis

plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

font1 = {'family': 'serif', 'color': 'green', 'size': 15}
font2 = {'family': 'serif', 'color': 'red', 'size': 10}

plt.title("Comany Sales vs Rate", fontdict=font1)
plt.xlabel("company rate", fontdict=font2)
plt.ylabel("company sales", fontdict=font2)

plt.plot(x, y)

plt.grid(axis='y')
# show only 'y' axis

plt.show()
```

### **Line Properties**

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

plt.title("Comany Sales vs Rate")
plt.xlabel("company rate")
plt.ylabel("company sales")

plt.plot(x, y)

plt.grid(color='green', linestyle='--', linewidth=0.5)

plt.show()
```

## `Display Multiple Plots`

>##### Here, we are using `subplot()` function for draw multiple plots in one figure.

```python
# plot 1:
x = np.array([1, 6, 3, 13, 5])
y = np.array([2, 5, 20, 7, 11])

plt.subplot(1, 2, 1)
plt.plot(x, y)

# plot 2:
x = np.array([0, 1, 2, 3, 4])
y = np.array([2, 5, 20, 7, 11])

plt.subplot(1, 2, 2)
plt.plot(x, y)

plt.show()
```
