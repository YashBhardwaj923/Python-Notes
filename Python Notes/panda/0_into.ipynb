# Introduction to `NumPy`, `Pandas`, `Matplotlib`.

## 1. NumPy (Numerical Python) :

> #### **Purpose**: NumPy is the foundational package for numerical and array-based computations in Python. It provides support for multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these data structures.

```python
import numpy as np
```

```python
arr = [1, 4, 6, 8, 10, 12] # list
array = np.array(arr) # array
print(array * 2)
```

## 2. Pandas (Panel Data) :

> #### **Purpose**: Pandas is designed for data manipulation and analysis. It provides powerful tools for handling structured data such as tabular data (like SQL tables or Excel spreadsheets)

```python
import pandas as pd
```

```python
lst = ["Rohan", "Sohan", "Anuj", "Gopal", "Ayushi", "Madhu"]

df = pd.DataFrame(lst)
print(df)
```

## 3. Matplotlib :

> #### **Purpose**: Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It serves as a fundamental plotting library for various types of data visualizations.

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 6, 12, 12]
y = [10, 20, 30, 40, 50]
```

```python
plt.plot(x, y)
plt.title("Line Graph with Matplotlib")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#### These libraries are often used together in data analysis workflows:

* > Use `NumPy` for numerical computations.
* > Use `Pandas` to handle and preprocess data.
* > Use `Matplotlib` to visualize the processed data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create data with numpy
data = np.random.randn(100, 3)

# convert to a pandas dataframe
df = pd.DataFrame(data, columns=['A', 'B', 'C'])
```

```python
# plot a histogram for column 'B'
plt.hist(df['B'], bins=20, alpha=0.9)
plt.title('Histogram of Column B')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
