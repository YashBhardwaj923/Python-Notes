# What is Pandas?

> * Pandas is a Python library used for working with **data series**.

> * It has functions for analyzing, cleaning, exploring and manipulating data.

> * The name `Pandas` has a reference to both `Panel Data` and `Python Data Analysis` and was created by *Wes McKinney* in 2008.

# What can Pandas Do?

> Pandas gives you answer about the data. Like- `Is there a correlation between two or more columns?`, `What is average value?`, `Max value?`, `Min value?`. Pandas are also able to delete rows that are not relevant, or contains wrong values, like empty or NULL values. This called `cleaning` the data.

# ***What is a Series?***

A Pandas Series is like a column in a table. It is a one-dimensional array holding data of any type.

```python
import pandas as pd
```

```python
a = [1, 8, 9, 20]

myvar = pd.Series(a)
print(myvar)
```

```python
numData = [10, 20, 40, 50, 60]

serData = pd.Series(numData)
print(serData)
```

### ***Lables?***

If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc. This label can be used to access a specified value.

```python
print(myvar[2])
```

```python
print(serData[3])
```

### ***Create Labels***

```python
arr = [10, 20, 30 ,40, 50]

ser = pd.Series(arr, index=["x", "y", "z", "m", "n"])
print(ser)

#? Note: When we have create labels, we can access an item by referring to the label.
```

```python
print(ser["m"])
```

```python
print(ser["z"])
```

```python
user_data = [1, "Mohan Kumar", 34, "New Delhi"]

sers = pd.Series(user_data, index=["id", "name", "age", "city"])
print(sers)
```

```python
print(sers["name"])
```

### ***Key/Value Object as Series***

```python
std_hours = {
    "day1": 10,
    "day2": 40,
    "day3": 30,
    "day4": 20
}

mydata = pd.Series(std_hours)
print(mydata)

#? The keys of the dictionary become the index of the Series. 
```

```python
mydata = pd.Series(std_hours, index=["day1", "day2"])
print(mydata)
```

```python
mydata = pd.Series(std_hours, index=["day3", "day4"])
print(mydata)
```

# ***What is DataFrames?***

> A Pandas DataFrame is a 2 dimansional data structure, like a 2 dimansional array or a table with rows and columns.

```python
data = {
    "id": [1, 2, 3, 4, 5],
    "name": ["Rahul", "Rohit", "Raj", "Ravi", "Ramesh"],
    "marks": [50, 60, 70, 80, 90]
}

df = pd.DataFrame(data)
print(df)
```

### ***Locate Row***

As you can see from the result above, the DataFrame is like a table with rows and columns. Pandas use the `loc` attribute to return one or more specified row or rows.

```python
print(df.loc[3])
```

```python
print(df.loc[[0, 1, 3]])

#? When using `[]`, the result is Pandas DataFrame. 
```

### ***Named Indexes***

```python
df = pd.DataFrame(data, index=["a", "b", "c", "d", "e"])
print(df)
```

```python
print(df.loc[["b", "c"]])
```

# ***Analyzing DataFrames***

### ***Viewing the Data***

#### **`head()`**

> ##### One of the most used method for getting a quick overview of the DataFrame, is the `head()` method. The `head()` method returns the headers and a specified number of rows, starting from the top.

```python
import pandas as pd
```

```python
df = pd.read_csv("hotel_bookings.csv")

cl_data = df.head(20)

print(cl_data.to_string())
```

#### **`tail()`**

> ##### There is also a `tail()` method for viewing the last rows of the DataFrames. Starting fomr the bottom.

```python
cl_data = df.tail(20)

print(cl_data.to_string())
```

#### **`info()`**

> ##### The DataFrame object has a method called `info()`, that gives you more information about the data set.

```python
print(df.info())
```
