```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random sales data
dates = pd.date_range(start="2024-01-01", periods=30)  # 30 days of data
products = ['Laptop', 'Smartphone', 'Tablet', 'Headphones', 'Smartwatch']

# Create a DataFrame with random sales numbers
np.random.seed(42)  # For reproducibility
data = {
    'Date': np.tile(dates, len(products)),
    'Product': np.repeat(products, len(dates)),
    'Sales': np.random.randint(10, 100, size=len(dates) * len(products))
}
df = pd.DataFrame(data)

# Aggregate total sales per product
total_sales = df.groupby('Product')['Sales'].sum().reset_index()

# Plot sales trends
plt.figure(figsize=(10, 5))
for product in products:
    subset = df[df['Product'] == product]
    plt.plot(subset['Date'], subset['Sales'], marker='o', label=product)

plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Daily Sales Trends by Product')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Print total sales
print("Total Sales by Product:")
print(total_sales)
```
