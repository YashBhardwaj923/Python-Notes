```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load real company sales data (Assume 'company_sales.csv' is provided)
df = pd.read_csv('company_sales.csv')

# Display basic info and preview data
print("Data Summary:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Aggregate total sales per product
total_sales = df.groupby('Product')['Sales'].sum().reset_index()

# Plot sales trends
plt.figure(figsize=(10, 5))
products = df['Product'].unique()
for product in products:
    subset = df[df['Product'] == product]
    plt.plot(subset['Date'], subset['Sales'], marker='o', label=product)

plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Company Sales Trends by Product')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Print total sales
print("Total Sales by Product:")
print(total_sales)
```
